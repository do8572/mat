```python
import pandas as pd, numpy as np
from math import isnan
```

```python
tar = "binom"

ess = pd.read_csv("table/ess- "+tar+" .csv")
ess["mse"] = ess.x1 + ess.x2
ess = ess.sort_values("mse")

esssec = pd.read_csv("table/esssec- "+tar+" .csv")
esssec["mse"] = esssec.x1 + esssec.x2
esssec = esssec.sort_values("mse")

for (i1, row), (i2, row2) in zip(ess.iterrows(), esssec.iterrows()):
    print(f"{row.alg}{'(T)' if row.tune else ''} & ${round(row.x1) if not isnan(row.x1) else '-'} \\pm {round(row['x1.1']) if not isnan(row['x1.1']) else '-'}$ & ${round(row.x2) if not isnan(row.x2) else '-'} \\pm {round(row['x2.1']) if not isnan(row['x2.1']) else '-'}$\\\\")
    
print(20 * "--")
    
for (i1, row), (i2, row2) in zip(ess.iterrows(), esssec.iterrows()):
    print(f"{row2.alg}{'(T)' if row2.tune else ''} & ${round(row2.x1) if not isnan(row2.x1) else '-'} \\pm {round(row2['x1.1']) if not isnan(row2.x1) else '-'}$ & ${round(row2.x2) if not isnan(row2.x1) else '-'} \\pm {round(row2['x2.1']) if not isnan(row2.x1) else '-'}$\\\\")

print(20 * "--")
    
for (i1, row), (i2, row2) in zip(ess.iterrows(), esssec.iterrows()):
    print(f"{row.alg}{'(T)' if row.tune else ''} & ${round(row.x1) if not isnan(row.x1) else '-'} \\pm {round(row['x1.1']) if not isnan(row['x1.1']) else '-'}$ & ${round(row.x2) if not isnan(row.x2) else '-'} \\pm {round(row['x2.1']) if not isnan(row['x2.1']) else '-'}$ & ${round(row2.x1) if not isnan(row2.x1) else '-'} \\pm {round(row2['x1.1']) if not isnan(row2['x1.1']) else '-'}$ & ${round(row2.x2) if not isnan(row2.x2) else '-'} \\pm {round(row2['x2.1']) if not isnan(row2['x2.1']) else '-'}$\\\\")
```

```python
tar = "binom_full"

ess = pd.read_csv("table/ess- "+tar+" .csv")
ess["mse"] = ess.x1 + ess.x2
ess = ess.sort_values("mse")

esssec = pd.read_csv("table/esssec- "+tar+" .csv")
esssec["mse"] = esssec.x1 + esssec.x2
esssec = esssec.sort_values("mse")

for (i1, row), (i2, row2) in zip(ess.iterrows(), esssec.iterrows()):
    print(f"{row.alg}{'(T)' if row.tune else ''} & ${round(row.x1) if not isnan(row.x1) else '-'} \\pm {round(row['x1.1']) if not isnan(row['x1.1']) else '-'}$ & ${round(row.x2) if not isnan(row.x2) else '-'} \\pm {round(row['x2.1']) if not isnan(row['x2.1']) else '-'}$\\\\")
    
print(20 * "--")
    
for (i1, row), (i2, row2) in zip(ess.iterrows(), esssec.iterrows()):
    print(f"{row2.alg}{'(T)' if row2.tune else ''} & ${round(row2.x1) if not isnan(row2.x1) else '-'} \\pm {round(row2['x1.1']) if not isnan(row2.x1) else '-'}$ & ${round(row2.x2) if not isnan(row2.x1) else '-'} \\pm {round(row2['x2.1']) if not isnan(row2.x1) else '-'}$\\\\")

print(20 * "--")
    
for (i1, row), (i2, row2) in zip(ess.iterrows(), esssec.iterrows()):
    print(f"{row.alg}{'(T)' if row.tune else ''} & ${round(row.x1) if not isnan(row.x1) else '-'} \\pm {round(row['x1.1']) if not isnan(row['x1.1']) else '-'}$ & ${round(row.x2) if not isnan(row.x2) else '-'} \\pm {round(row['x2.1']) if not isnan(row['x2.1']) else '-'}$ & ${round(row2.x1) if not isnan(row2.x1) else '-'} \\pm {round(row2['x1.1']) if not isnan(row2['x1.1']) else '-'}$ & ${round(row2.x2) if not isnan(row2.x2) else '-'} \\pm {round(row2['x2.1']) if not isnan(row2['x2.1']) else '-'}$\\\\")

ess["x_avg"] = ess[["x" + str(i) for i in range(1, 12)]].mean(axis=1)
ess["x_std"] = ess[["x" + str(i) for i in range(1, 12)]].std(axis=1)
ess["mse"] = ess["x_avg"] - 2 * ess["x_std"]
ess = ess.sort_values("mse")

esssec["x_avg"] = esssec[["x" + str(i) for i in range(1, 12)]].mean(axis=1)
esssec["x_std"] = esssec[["x" + str(i) for i in range(1, 12)]].std(axis=1)
esssec["mse"] = esssec["x_avg"] - 2 * esssec["x_std"]
esssec = esssec.sort_values("mse")

print(20 * "--")

for (i1, row), (i2, row2) in zip(ess.iterrows(), esssec.iterrows()):
    print(f"{row.alg}{'(T)' if row.tune else ''} & ${round(row.x_avg)} \\pm {round(row.x_std)}$ & ${round(row2.x_avg)} \\pm {round(row2.x_std)}$\\\\")

```
